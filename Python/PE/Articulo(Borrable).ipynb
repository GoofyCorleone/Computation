```python
import sympy as sp
import numpy as np
from sympy import I
from sympy.physics.quantum import TensorProduct

alpha1 , chi1 , delta  = sp.symbols('\\alpha_1 \\chi_1 \\delta', real=True)
pi = sp.pi
theta1 , theta2 , p , q = sp.symbols('\\theta_1 \\theta_2 p q',real=True)
vecE = sp.Matrix(
    [
        sp.cos(chi1),
        I*sp.sin(chi1)
    ]
)
rotalpha = sp.Matrix(
    [
        [sp.cos(alpha1) , -sp.sin(alpha1)],
        [sp.sin(alpha1) , sp.cos(alpha1)]
    ]
)
vecE = rotalpha * vecE
vecE
```

```python
alpha2 , chi2 , delta  = sp.symbols('\\alpha_2 \\chi_2 \\delta', real=True)
pi = sp.pi
theta1 , theta2 , p , q = sp.symbols('\\theta_1 \\theta_2 p q',real=True)
vecE2 = sp.Matrix(
    [
        sp.cos(chi2),
        I*sp.sin(chi2)
    ]
)
rotalpha = sp.Matrix(
    [
        [sp.cos(alpha2) , -sp.sin(alpha2)],
        [sp.sin(alpha2) , sp.cos(alpha2)]
    ]
)
vecE2 = rotalpha * vecE2
proy = vecE2*vecE2.adjoint()
proy = sp.simplify(proy)
proy
```
