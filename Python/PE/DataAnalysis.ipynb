```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt

# names = ['DATICOS/PElip/chi{}.txt'.format(i) for i in np.arange(0,370,10)]
names = ['DatosPE/chi{}.txt'.format(i) for i in np.arange(0,365,5)]
# DataCounts = np.zeros((37,5000))
DataCounts = np.zeros((74,5000))
DataI = pd.read_csv('DatosPE/incidente pl4.txt',header=19, delimiter='\t')
intensidad = DataI['Counts per Bin '].values
def gettinData(route):
    Data = pd.read_csv(route, header=19, delimiter='\t')
    Counts = Data['Counts per Bin '].values
    return Counts

def saveData(names):
    for i in range(len(names)):
        Counts = gettinData(route=names[i])
        DataCounts[i,0:len(Counts)] = Counts
        
def C12(alpha,chi,alphap,chip):
    t1 = ( np.sin(alpha-alphap) * np.sin(chi+chip) )**2
    t2 = ( np.cos(alpha-alphap) * np.cos(chi-chip) )**2
    return t1 + t2

saveData(names=names)
DataCounts = DataCounts.T
Nmean = []

# for i in range(len(names)):
#     aux = DataCounts[0:2000,i]
#     if i in [0,1,2]:
#         Nmean.append(aux.mean())
#     else:
#         Nmean.append(aux.mean())

for i in range(len(names)):
    aux = DataCounts[:,i] / intensidad
    Nmean.append(aux.mean())
Nmean = np.array(Nmean)
Nmean
```

```python
x = np.arange(0,365,5) * np.pi / 180
xarr = np.linspace(0,365,200) * np.pi / 180
y = C12(alpha=-3.31,chi=1.17,alphap=0,chip=-xarr+np.pi/3)
plt.scatter(x,Nmean/max(Nmean),c='r')
# plt.plot(x,Nmean/max(Nmean))
plt.plot(xarr,y)
```
