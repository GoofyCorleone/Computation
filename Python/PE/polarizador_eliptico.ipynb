```python
import sympy as spy
```

```python
alpha, chi, delta, theta1, theta2 = spy.symbols('\\alpha, \chi, \delta, \\theta_1, \\theta_2', real = True)

E1 = spy.cos(alpha)*spy.cos(chi) - spy.I*spy.sin(alpha)*spy.sin(chi)
E2 = spy.sin(alpha)*spy.cos(chi) + spy.I*spy.cos(alpha)*spy.sin(chi)

e = spy.Matrix([[E1],[E2]])
p = e*e.adjoint()

R1 = spy.Matrix([[spy.cos(theta1), spy.sin(theta1)],
               [-spy.sin(theta1), spy.cos(theta1)]])

R1p = spy.Matrix([[spy.cos(theta1 + spy.pi/2), spy.sin(theta1 + spy.pi/2)],
               [-spy.sin(theta1 + spy.pi/2), spy.cos(theta1 + spy.pi/2)]])

R2 = spy.Matrix([[spy.cos(theta2), spy.sin(theta2)],
               [-spy.sin(theta2), spy.cos(theta2)]])

Q = spy.Matrix([[spy.exp(spy.I*spy.pi/2), 0],
               [0, 1]])

Q1 = R1.transpose()*Q*R1

pl = spy.Matrix([[spy.cos(theta2)**2, spy.cos(theta2)*spy.sin(theta2)],
               [spy.cos(theta2)*spy.sin(theta2), spy.sin(theta2)**2]])

Q1p = R1p.transpose()*Q*R1p

p2 =Q1p*pl*Q1
p2.simplify()
p2
```

```python
Pr = spy.expand(p2)
Pr 
```

```python
Pr = Pr.subs(theta1,alpha)
Pr = Pr.subs(theta2,alpha-chi)
Pr
```

```python
PrT = spy.simplify(p)
PrT
```

```python
state = spy.Matrix([1,0])
Pr*state
```
