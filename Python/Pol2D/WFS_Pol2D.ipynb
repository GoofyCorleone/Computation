```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
routes = ['Datos/WVS00.csv','Datos/WVS450.csv','Datos/WVS900.csv','Datos/WVS4590.csv']
```

```python
Data_00 = pd.read_csv(routes[0],delimiter='\t')
Data_00.reset_index(inplace=True)

Data_450 = pd.read_csv(routes[1],delimiter='\t')
Data_450.reset_index(inplace=True)

Data_900 = pd.read_csv(routes[2],delimiter='\t')
Data_900.reset_index(inplace=True)

Data_4590 = pd.read_csv(routes[3],delimiter='\t')
Data_4590.reset_index(inplace=True)
```

```python
indices = ['level_{}'.format(i) for i in range(0,4)]
indices2=['{}'.format(i) for i in range(1,17)]
indices = indices + indices2
M_00 = np.zeros((len(Data_00[indices[0]].values),len(indices)))
M_450 = np.zeros((len(Data_00[indices[0]].values),len(indices)))
M_900 = np.zeros((len(Data_00[indices[0]].values),len(indices)))
M_4590 = np.zeros((len(Data_00[indices[0]].values),len(indices)))
```

```python
def MatrixFormation(Data,indices,M=np.array([])):
    for j in range (len(indices)):
        M[:,j] = Data[indices[j]].values
def StokesCalculation(I_00,I_450,I_900,I_4590):
    S_0 = I_00 + I_900
    S_1 = I_00 - I_900
    S_2 = 2*I_450 - I_00 - I_900
    S_3 = 2*I_4590 - I_00 - I_900
    return S_0 , S_1 , S_2 , S_3
```

```python
MatrixFormation(M=M_00,Data = Data_00 , indices = indices)
MatrixFormation(M=M_450,Data = Data_450 , indices = indices)
MatrixFormation(M=M_900,Data = Data_900 , indices = indices)
MatrixFormation(M=M_4590,Data = Data_4590 , indices = indices)
```

```python
S0 , S1 , S2 , S3= StokesCalculation(M_00,M_450,M_900,M_4590)
```

```python
s0 , s1 , s2 , s3 = S0.mean() , S1.mean() , S2.mean() , S3.mean()
s0 , s1 , s2 , s3 =  s0/s0 , s1/s0 , s2/s0 , s3/s0
print(s0,s1,s2,s3)
```
