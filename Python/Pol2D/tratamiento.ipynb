```python
import cv2
import matplotlib.pyplot as plt

# Lee la imagen en modo de escala de grises
imagen_gris = cv2.imread('Datos/MapaObjeto_0(101).tif', cv2.IMREAD_GRAYSCALE)

# Mostrar la imagen en el Jupyter Notebook
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen en Escala de Grises con OpenCV')
plt.axis('off')  # Para ocultar los ejes
plt.show()

# Imprimir las intensidades de los píxeles
print("Intensidades de los píxeles:\n", imagen_gris)
print(max(imagen_gris))
```

```python
imagen_cv2 = cv2.imread('Datos/MapaObjeto_0(101).tif', cv2.IMREAD_GRAYSCALE)
cv2.imshow('Imagen en Escala de Grises con OpenCV', imagen_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
print("Intensidades con OpenCV:\n", imagen_cv2)
```
