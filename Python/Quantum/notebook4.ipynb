```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.x(0)
qc.draw('mpl')
```

```python
q = Statevector(qc)
q.draw('latex')
```

```python
counts_sv = q.sample_counts(100)
print(counts_sv)
```

```python
qc.measure_all()
qc.draw('mpl')
```

```python
from qiskit.providers.basic_provider import BasicSimulator
sim = BasicSimulator()
```

```python
counts_sim = sim.run(qc, shots=100).result().get_counts()
print(counts_sim)
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(counts_sim)
```
