```python
import sympy as sp
import numpy as np
from sympy import I
```

```python
Ex , Ey = sp.symbols('Ex Ey', type=complex)
g11 , g12 , g21 , g22 = sp.symbols('Gamma_11 Gamma_12 Gamma_21 Gamma_22',type=complex)
tau = sp.symbols('\\tau')
gammaP = sp.Function('\gamma', real=False)(tau)
gammaN = sp.Function('\gamma', real=False)(-tau)

g11 = Ex*sp.conjugate(Ex)
g12 = Ex*sp.conjugate(Ey)*gammaP
g21 = Ey*sp.conjugate(Ex)*gammaN
g22 = Ey*sp.conjugate(Ey)

Gamma = sp.Matrix(
    [
        [g11 , g12],
        [g21, g22]
    ])

def Jones2Stokes(Gamma,Cov=False):
    s0 , s1 , s2 , s3 = sp.symbols('S0 S1 S2 S3')
    s0 = Gamma[0,0] + Gamma[1,1]
    s1 = Gamma[0,0] - Gamma[1,1]
    s2 = Gamma[1,0] + Gamma[0,1]
    s3 = I * ( Gamma[1,0] - Gamma[0,1] )
    S = sp.Matrix([s0,s1,s2,s3])

    if Cov:
        M = sp.eye(4)
        for i in range(0,4):
            for j in range(0,4):
                M[i,j] =sp.Symbol(f'<{S[i]}*{S[j]}>') - sp.Symbol(f'<{S[i]}>*<{S[j]}>')
        return S , M
    else:
        return S
    
S , M= Jones2Stokes(Gamma,True)
M[1,1]
```
